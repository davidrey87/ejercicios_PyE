```python
#Importamos las librerias necesarias
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
```

```python
#Abrimos el archivo con los datos historicos con los que crearemos nuestro modelo
X_historico = pd.read_csv('historico.csv')
```

```python
#Removemos los datos categoricos para convertilos a datos numericos, de esta manera la 
#libreria sklearn los podra interpretar
Y_historico = pd.get_dummies(X_historico.pop("especie"))
```

```python
#Creamos nuestro segmento de entrenamiento y nuestro segmento de pruebas equivalente 
#al 80% - 20%
X_train = X_historico[0:80]
X_test = X_historico[80:100]
Y_train = Y_historico[0:80]
Y_test = Y_historico[80:100]
```

```python
#Definimos un arbol binario
arbol = DecisionTreeClassifier(random_state = 0, max_depth=2)
```

```python
#Entrenamos nuestro arbol
arbol.fit(X_train,Y_train)
```

```python
#Abrimos el archivo de sirenas a predecir
X_sirenas = pd.read_csv('sirenas.csv')
```

```python
#Lo preparamos removiento lo que sera en un futuro la columna de datos categoricos
Y_sirenas = X_sirenas.pop("especie")
```

```python
#Mandamos los datos a predecir colocando la cabecera para los datos numericos de sierenas endemicas y/o migrantes
Y_sirenas = pd.DataFrame(data = arbol.predict(X_sirenas), columns = ['sirena_endemica','sirena_migrante'])
```

```python
#Convertimos los datos numericos en categoricos
Y_sirenas["especie"] = np.where(Y_sirenas["sirena_endemica"] == 1, 'sirena_endemica', 'sirena_migrante')
```

```python
#Colocamos los datos categoricos en el conjunto de datos que enviaremos a un CVS
X_sirenas["especie"] = Y_sirenas.pop("especie")
```

```python
X_sirenas.to_csv('sirenas_predict.csv', index=False)
```
