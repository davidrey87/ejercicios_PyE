```python
dress.dropna(subset=['SleeveLength'], how='all', inplace = True)
SleeveLength = pd.Categorical(dress["SleeveLength"])
SleeveLength.describe()
```

```python

%matplotlib inline
from sklearn.decomposition import PCA 
from sklearn.cluster import DBSCAN

pca = PCA(n_components=2)
pData = pca.fit_transform(datos)
print(pData.shape)
```

```python

import matplotlib.pyplot as plt

color = np.array(['red','blue'])

data2 = pd.DataFrame(pData)
plt.scatter(data2[0],data2[1], c=color)
plt.show()
```

```python
pca.components_
```

```python
est = DBSCAN(eps=0.6, min_samples = 10, algorithm='kd_tree', metric='euclidean')
est.fit(data2)
labels = est.labels_
```

```python
print(labels)
```

```python
from collections import Counter
print(Counter(labels))
```

```python
def set_colors(labels, colors='rgmyrggkyrgb'):
    colored_labels = []
    for label in labels:
        colored_labels.append(colors[label])
    return colored_labels
```

```python
colores = set_colors(labels)
plt.scatter(data2[0],data2[1], c=colores)
plt.show()
```
