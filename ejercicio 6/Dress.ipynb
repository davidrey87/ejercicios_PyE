```python
import pandas as pd
import numpy as np
```

```python
dress = pd.read_excel("dress.xlsx")
dress.head()
```

```python
style = pd.Categorical(dress["Style"])
style.describe()
```

```python
dress['Style']= dress['Style'].str.lower()
```

```python
dress["Style"] = dress["Style"].astype("category").cat.codes
style = pd.Categorical(dress["Style"])
style.describe()
```

```python
dress.dropna(subset=['Season'], how='all', inplace = True)
season = pd.Categorical(dress["Season"])
season.describe()
```

```python
dress['Season']= dress['Season'].str.lower()
```

```python
season = pd.Categorical(dress["Season"])
season.describe()
```

```python
dress["Season"].replace('autumn','automn', inplace=True)
```

```python
season = pd.Categorical(dress["Season"])
season.describe()
```

```python
dress["Season"] = dress["Season"].astype("category").cat.codes
season = pd.Categorical(dress["Season"])
season.describe()
```

```python
dress.dropna(subset=['SleeveLength'], how='all', inplace = True)
SleeveLength = pd.Categorical(dress["SleeveLength"])
SleeveLength.describe()
```

```python
dress["SleeveLength"].replace('cap-sleeves','capsleeves', inplace=True)
dress["SleeveLength"].replace('sleeevless','sleeveless', inplace=True)
dress["SleeveLength"].replace('sleevless','sleeveless', inplace=True)
dress["SleeveLength"].replace('sleveless','sleeveless', inplace=True)
dress["SleeveLength"].replace('threequater','threequarter', inplace=True)
dress["SleeveLength"].replace('urndowncollor','turndowncollor', inplace=True)
dress['SleeveLength']= dress['SleeveLength'].str.lower()
```

```python
dress["SleeveLength"] = dress["SleeveLength"].astype("category").cat.codes
SleeveLength = pd.Categorical(dress["SleeveLength"])
SleeveLength.describe()
```

```python
dress.dropna(subset=['Price'], how='all', inplace = True)
price = pd.Categorical(dress["Price"])
price.describe()
```

```python
dress["Price"].replace('Avarage','Medium', inplace=True)
dress['Price']= dress['Price'].str.lower()
```

```python
price = pd.Categorical(dress["Price"])
price.describe()
```

```python
dress["Price"] = dress["Price"].astype("category").cat.codes
```

```python
price = pd.Categorical(dress["Price"])
price.describe()
```
