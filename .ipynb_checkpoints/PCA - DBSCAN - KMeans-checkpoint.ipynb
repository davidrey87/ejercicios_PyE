```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.decomposition import PCA
```

```python
datos = pd.read_csv('iris.csv', header=None)
```

```python
target = datos.pop(4) 
```

```python
target.head()
```

```python
pca = PCA(n_components=2)
pData = pca.fit_transform(datos)
pData.shape
```

```python
dt = pd.DataFrame(pData)
colores = np.array(['black','red','blue','lime'])
plt.scatter(dt[0],dt[1], c=colores[target])
plt.show()
```

```python
pca.components_
```

```python
from sklearn.cluster import DBSCAN
from collections import Counter
```

```python
#'auto', 'brute','kd_tree', 'ball_tree'

estimador = DBSCAN(eps=0.55,
                  min_samples = 10,
                  algorithm='auto',
                  metric = 'euclidean')
estimador.fit(dt)

labels = estimador.labels_
print Counter(labels)
```

```python
def set_colors(labels, colors='rgbykcm'):
    colored_labels = []
    for label in labels:
        colored_labels.append(colors[label])
    return colored_labels
```

```python
colores = set_colors(labels)
plt.scatter(dt[0],dt[1], c=colores)
plt.xlabel("x")
plt.xlabel("y")
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=3)
model.fit(dt)
```

```python
model.labels_
```

```python
colores = np.array(['black','red','blue','lime'])
plt.scatter(dt[0],dt[1], c=colores[model.labels_])
plt.show()
```

```python
import sklearn.metrics as sm
```

```python
sm.accuracy_score(target,model.labels_)
```

```python
matrix = sm.confusion_matrix(target,model.labels_)
matrix
```
