```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn import cross_validation
```

```python
data = pd.read_csv("tic-tac-toe.data", header=None)
```

```python
data.head()
```

```python
#Convertiremos los datos categoricos a numericos X 2 = x, 1 = o y 0 = b. Y Positive = 1, Negative = 0
funcion = preprocessing.LabelEncoder()
for f in data:
    data[f]=funcion.fit_transform(data[f])
data.head()
```
