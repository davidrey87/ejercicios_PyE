```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn.decomposition import PCA
```

```python
st = pd.read_csv("calidad_del_aire_cyl_1997_2013.csv", delimiter=";", index_col = "DIA", parse_dates = True)
st.head()
```

```python
headers = list(st)
headers.remove("PROVINCIA")
headers.remove("ESTACIÓN")
headers
```

```python
for i in headers:
    gas = pd.Categorical(st[i])
    print(gas.describe())
```

```python
#for i in headers:
#    st[i] = st[i].fillna(0)
#    st[i] = st[i].replace("No cumple el anexo IV de la Decisión del Consejo 97/101/CE",0)
#    st[i] = st[i].astype(float)
#st.head()
st.dropna(subset=['NO2 (ug/m3)'], how='all', inplace = True)
st = st[(st['NO2 (ug/m3)'] != "No cumple el anexo IV de la Decisión del Consejo 97/101/CE")]
#st = st[(st['NO2 (ug/m3)'] != 0)]
st['NO2 (ug/m3)'] = st['NO2 (ug/m3)'].astype(float)
```

```python
for i in headers:
    gas = pd.Categorical(st[i])
    print(gas.describe())
```

```python
#Aplicamos un smooth a los datos por año
tsd = st['NO2 (ug/m3)'].groupby(pd.Grouper(freq='M')).agg('mean')
```

```python
zonas = pd.Categorical(st['PROVINCIA']+';'+st['ESTACIÓN'])
zonas_unicos = zonas.categories.unique()
print(zonas_unicos)
```

```python
#dtt = st[(st['PROVINCIA'] == 'BURGOS') & (st['ESTACIÓN'] == 'Burgos 5')]
#dtt
from sklearn.preprocessing import StandardScaler
escala = StandardScaler()
```

```python
for i in zonas_unicos:
    x = i.split(';')
    print(x[0]+" - "+x[1])
    dtt = st[(st['PROVINCIA'] == x[0]) & (st['ESTACIÓN'] == x[1])]
    print(dtt['NO2 (ug/m3)'])
    dtt.pop('PROVINCIA')
    dtt.pop('ESTACIÓN')
    tsd = dtt['NO2 (ug/m3)'].groupby(pd.Grouper(freq='M')).agg('mean')

    print(sm.stats.durbin_watson(tsd))
    model = ARIMA(tsd, order=(1,0,0))
    model_fit = model.fit()
    pred = model_fit.predict(start=tsd.shape[0]-2, end=tsd.shape[0]+2)
    plt.plot(tsd)
    plt.plot(pred)
    print(plt.show())
```
